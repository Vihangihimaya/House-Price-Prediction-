Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the Boston House Price Dataset

```python
house_price_dataset = sklearn.datasets.load_boston()
```

```python
print(house_price_dataset)
```

```python
# Loading the dataset  to a pandas dataframe
house_price_dataframe = pd.DataFrame(house_price_dataset.data)
```

```python
# Print First 5 rows of our Dataframe
house_price_dataframe.head()
```

```python
# add the target (price ) column to the Datafreame
house_price_dataframe['price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
# checking the number of rows and Columns in the data frame
house_price_dataframe.shape
```

```python
#check for missing vallues
house_price_dataframe.isnull().sum()
```

```python
# Statical measures of the dataset
house_price_dataframe.describe()
```

Understanding the correlation between various features in the dataset

1. positive correlation
2. negative correlation

```python
correlation = house_price_dataframe.corr()
```

```python
# constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True, fmt= ',.1f' , annot=True, cmap='Blues')
```

Splitting the data and target

```python
X = house_price_dataframe.drop(['price'], axis=1)
Y = house_price_dataframe['price']
```

```python
print(X)
print(Y)
```

Splitting the data into Training data test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGBoost Regressor

```python
# loading the model
model = XGBRegressor()
```

```python
# training the model with X_train
model.fit(X_train, Y_train)
```

Evaluation

Prediction on training data

```python
# accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absplutr Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolate Error :', score_2)
```

Visualizing the actual Price and prediction prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs predicted Prices")
plt.show()
```

Prediction  on test data

```python
# accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error ; " , score_1)
print('Mean Absolute Error : ', score_2)
```
